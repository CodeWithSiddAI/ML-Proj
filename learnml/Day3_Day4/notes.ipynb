### Lecture 1: The Formal Cost Function $J(w, b)$

In ML, we need a mathematical scoreboard to tell us how badly our model is doing. We call this the **Cost Function**, denoted by the letter $J$.

We use **Mean Squared Error (MSE)**. You already saw a simplified version of this, but here is the formal mathematical equation Andrew Ng teaches. Write this down exactly as it appears:

$$J(w,b) = \frac{1}{2m} \sum_{i=1}^{m} (\hat{y}^{(i)} - y^{(i)})^2$$

Let's break down this Greek alphabet soup using the Feynman technique:

* $m$: The total number of houses in your dataset.
* $\sum$ (Sigma): A giant loop. It means "Do the math for the first house, then the second, then the third... and add all the results together."
* $i$: The index. $y^{(1)}$ means the price of the 1st house. $y^{(5)}$ is the 5th house.
* $\hat{y} - y$: (Our Guess) minus (The Actual Price). This is our raw error.
* $^2$: We square the error so negative and positive errors don't cancel each other out, and to deeply punish the model for being *really* wrong.

**The "Aha!" Moment:** Why is it divided by $2m$ instead of just $m$ (which would be a normal average)?
The $2$ is a mathematical parlor trick. In calculus, when you take the derivative of something squared, the $2$ drops down to the front. By putting a $2$ in the denominator now, they cancel out later ($\frac{2}{2} = 1$), making the final math incredibly clean.

### Lecture 2: Calculus - What is a Derivative?

If the Cost Function $J(w,b)$ is a giant 3D bowl, our goal is to find the bottom of the bowl where the error is zero.
But the computer is blind. It drops a random guess for $w$ and $b$ onto the bowl. How does it know which way to walk?

It uses **Derivatives**.
Forget everything you think you know about calculus. A derivative is simply a **Sensitivity Meter**.

* If I nudge the weight ($w$) up by $0.001$, how much does the Cost ($J$) explode or shrink?
* If the Cost *shrinks* when I nudge $w$ up, it means I am walking downhill. Good! Keep increasing $w$.
* If the Cost *explodes* when I nudge $w$ up, it means I am walking uphill. Bad! I need to decrease $w$.

### Lecture 3: Partial Derivatives (The Gradients)

Because we have two variables ($w$ and $b$), we can't just take one derivative. We have to take a **Partial Derivative**.

* **Partial Derivative of $w$ (written as $\frac{\partial J}{\partial w}$):** We freeze $b$ completely (treat it like a boring constant number) and only measure how sensitive the Cost is to $w$.
* **Partial Derivative of $b$ (written as $\frac{\partial J}{\partial b}$):** We freeze $w$ completely and only measure how sensitive the Cost is to $b$.

Here are the final, solved derivative equations (the "slopes") derived using the Chain Rule. Write these down:

**The Slope of $w$:**


$$\frac{\partial J}{\partial w} = \frac{1}{m} \sum_{i=1}^{m} (\hat{y}^{(i)} - y^{(i)}) x^{(i)}$$

**The Slope of $b$:**


$$\frac{\partial J}{\partial b} = \frac{1}{m} \sum_{i=1}^{m} (\hat{y}^{(i)} - y^{(i)})$$

Notice how the $\frac{1}{2}$ disappeared? That's because the squared exponent dropped down and canceled it out! Also notice that the slope of $w$ is multiplied by $x^{(i)}$ at the end, while the slope of $b$ is not. This is because $w$ is attached to $x$ in our original equation ($\hat{y} = wx + b$), while $b$ stands alone.

````python
## Modify it for a fucntion definition

import numpy as np

# Your dataset
x_train = np.array([1, 2, 3])
y_train = np.array([300, 500, 800])
current_w = 100
current_b = 100

def compute_cost_and_gradients(x, y, w, b):
    m = len(x)
    
    # 1. Calculate predictions (y_hat) for all x
    y_hat = (w * x) + b 
    
    # 2. Calculate the Cost (J)
    # YOUR CODE HERE
    cost = (1/(2*m)) * np.sum((y_hat - y)**2)
    
    # 3. Calculate dj_dw (the slope of w)
    dj_dw = (1/m) * np.sum((y_hat - y) * x)
    
    # 4. Calculate dj_db (the slope of b)
    # YOUR CODE HERE
    dj_db = (1/m) * np.sum(y_hat - y)
    return cost, dj_dw, dj_db

# Run your function
cost, dj_dw, dj_db = compute_cost_and_gradients(x_train, y_train, current_w, current_b)

print(f"Cost: {cost}")
print(f"dj_dw: {dj_dw:.2f}")
print(f"dj_db: {dj_db:.2f}")
````
