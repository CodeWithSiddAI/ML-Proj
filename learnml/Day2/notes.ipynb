

### Day 2: The Language of Data (Linear Algebra & NumPy)

Yesterday, we talked about predicting a house's price using just *one* feature: Square Footage. The math was simple: $\hat{y} = wx + b$.

But in the real world, houses are priced on square footage, *and* number of bedrooms, *and* age, *and* zip code. Suddenly, we have 4 features. Our equation now looks like this:

$$\hat{y} = (w_1x_1) + (w_2x_2) + (w_3x_3) + (w_4x_4) + b$$

If a dataset has 1,000 features, writing out that equation would take hours. Furthermore, if you tell standard Python to calculate that using a `for` loop across a million houses, your computer will freeze because Python is simply too slow at looping.

We solve both the math problem and the speed problem using **Linear Algebra**.

#### Lecture 1: Vectors and Matrices (The Concept)

Linear Algebra sounds intimidating, but it is just a way of organizing numbers into boxes so we can move them all at the same time.

* **Scalar:** A single number. (e.g., `5`)
* **Vector:** A single list of numbers. Think of this as one single row or one single column in a spreadsheet. (e.g., `[2, 4, 6]`). A single house with 2,000 sq ft, 3 beds, and 10 years old is a vector: `[2000, 3, 10]`.
* **Matrix:** A 2D grid of numbers. Think of this as the *entire* spreadsheet. A list of 100 houses, each with 3 features, is a Matrix.

#### Lecture 2: The Dot Product (The Math)

How do we multiply multiple weights by multiple features all at once? We use an operation called the **Dot Product**.

Let's use a grocery store analogy.

* You are buying **Apples** and **Bananas**.
* Your **Quantities Vector** (Features, $x$) is: `[3 apples, 2 bananas]`
* Your **Prices Vector** (Weights, $w$) is: `[$4 per apple, $1 per banana]`

To find your total cost, you don't need a massive equation. You just take the Dot Product of the two vectors.
**The Rule:** Multiply the matching pairs, then add the results together.

1. First items: $3 \times 4 = 12$
2. Second items: $2 \times 1 = 2$
3. Add them: $12 + 2 = 14$

The Dot Product is 14! This is exactly how Neural Networks and Machine Learning models process hundreds of features instantly. They take the dot product of the Feature Vector and the Weight Vector, then add the Bias ($b$) at the very end.

#### Lecture 3: Enter NumPy (The Code)

Standard Python doesn't know how to do Dot Products. If you try to multiply two Python lists together (`[3, 2] * [4, 1]`), Python will crash and give you an error.

To do Linear Algebra, Data Scientists use a library called **NumPy** (Numerical Python). Under the hood, NumPy isn't written in Python; it is written in C++, which is incredibly fast. NumPy allows us to do math on massive datasets instantly without writing a single `for` loop. This is called **Vectorization**.

Here is how you do the grocery store math in NumPy:

```python
import numpy as np

# Create our vectors using np.array()
x = np.array([3, 2])
w = np.array([4, 1])

# Calculate the dot product
total_cost = np.dot(x, w)

print(total_cost) # Output: 14

```
