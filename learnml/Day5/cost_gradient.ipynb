```python
## implementing complete cost and gradient 

import numpy as np
# Your dataset
x_train = np.array([1, 2, 3])
y_train = np.array([300, 500, 800])
current_w = 100
current_b = 100

# learning rate - 
# epochs - 

def compute_cost_and_gradient(x, y, w, b):

    # m = size of of the dataset / how many houses are there in dataset
    m = len(x)
    # prediction based on current weight (w) and b (bias)
    y_hat = (w * x) + b
    cost = (1/ (2*m)) * np.sum((y_hat - y)**2) 

    # gradients dj_dw and dj_db
    dj_dw = (1/m) * np.sum((y_hat - y) * x)
    dj_db = (1/m) * np.sum((y_hat - y))
    return cost, dj_dw, dj_db


def gradient_decent(x, y, w_in, b_in, alpha, epochs):

    w = w_in
    b = b_in
    for i in range(epochs):
        
        # first calculate cost and gradient based on available or imagined w, b
        cost, dj_dw, dj_db = compute_cost_and_gradient(x, y, w, b)

        # print(f"cost: {cost},  current_weight: {w},  current_bias: {b}, dj_dw: {dj_dw}, dj_db: {dj_db}")
        if i % 10 == 0:
            print(f"Epoch {i:3}: Cost = {cost:8.2f} | w = {w:6.2f} | b = {b:6.2f}")
            

        # now update current value of w, b
        w = w - alpha * dj_dw 
        b = b - alpha * dj_db 
    
    return w, b 

# alpha - learning rate
# epochs - how many time we are taking that steps 1
learning_rate = 0.05
epochs = 100

print("\n--- Starts Training ---")
final_w, final_b = gradient_decent(x_train, y_train, current_w, current_b, learning_rate, epochs)
print("\n--- Training Complete ---")

print(f"Final model: y_hat = {final_w:.2f} x + {final_b:.2f}")

```
