### Lecture 1: The Blindfolded Hiker (Intuition)

You already built the "Eyes" of the machine (the Cost function that calculates the error) and the "Nervous System" (the Gradients that feel the slope).

Now, we build the "Legs".

Imagine you are blindfolded on a mountain (the Cost Bowl). You feel with your foot that the slope goes down to the right (the Gradient). You decide to take a step to the right.

But **how big** of a step do you take?

* If you take a tiny baby step, it will take you a million years to reach the bottom.
* If you take a massive leap, you might accidentally jump completely *over* the valley and land higher up on the opposite side of the mountain!

In Machine Learning, your step size is called the **Learning Rate**, represented by the Greek letter Alpha ($\alpha$).
The number of steps you decide to take before stopping is called **Epochs**.

### Lecture 2: The Update Rule (The Math)

Andrew Ng calls this the "Update Rule". We want to take our *current* weight, and update it to a *new*, better weight.

Write this formula down in your notebook:

$$w = w - \alpha \frac{\partial J}{\partial w}$$

$$b = b - \alpha \frac{\partial J}{\partial b}$$

Let's break down the Feynman logic of that **minus sign**:

* Remember your output from Day 4? Your slope (`dj_dw`) was **negative** (-566.67). A negative slope means the mountain is sloping downwards to the right.
* To get to the bottom, we need to move our weight *to the right* (increase it).
* Look at the math: $w - (\text{positive } \alpha \times \text{negative slope})$.
* Subtracting a negative number becomes a **positive addition**. The math perfectly forces $w$ to increase, moving it exactly where it needs to go!
